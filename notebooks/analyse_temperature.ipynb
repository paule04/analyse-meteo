```python
import csv
import statistics

def calculer_moyenne_ecart_type(nom_fichier, nom_colonne):
    """
    Calcule la moyenne et l'écart-type d'une colonne spécifiée dans un fichier CSV.

    Args:
        nom_fichier (str): Le chemin du fichier CSV.
        nom_colonne (str): Le nom de la colonne à analyser.

    Returns:
        tuple: Un tuple contenant la moyenne et l'écart-type, ou None si une erreur se produit.
    """

    try:
        with open(nom_fichier, 'r') as fichier_csv:
            lecteur_csv = csv.DictReader(fichier_csv)
            donnees = []
            for ligne in lecteur_csv:
                try:
                    donnees.append(float(ligne[nom_colonne]))
                except ValueError:
                    print(f"Erreur : La colonne '{nom_colonne}' contient des données non numériques.")
                    return None

        moyenne = statistics.mean(donnees)
        ecart_type = statistics.stdev(donnees)
        return moyenne, ecart_type

    except FileNotFoundError:
        print(f"Erreur : Le fichier '{nom_fichier}' n'existe pas.")
        return None
    except KeyError:
        print(f"Erreur : La colonne '{nom_colonne}' n'existe pas dans le fichier CSV.")
        return None

# Exemple d'utilisation
nom_fichier_csv = 'donnees.csv'  # Remplacez par le nom de votre fichier CSV
nom_colonne_a_analyser = 'temperature'  # Remplacez par le nom de la colonne à analyser

resultats = calculer_moyenne_ecart_type(nom_fichier_csv, nom_colonne_a_analyser)

if resultats:
    moyenne, ecart_type = resultats
    print(f"Moyenne de la colonne '{nom_colonne_a_analyser}': {moyenne}")
    print(f"Écart-type de la colonne '{nom_colonne_a_analyser}': {ecart_type}")
```
